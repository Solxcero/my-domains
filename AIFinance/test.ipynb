```python
import pandas as pd
import yfinance as yf
import pandas_datareader.data as pdr
import talib
```

```python
appl_data = yf.download("AAPL", start="2024-01-01", end="2024-12-31")

# 1. 단순이동평균(SAM), 20일 기준
appl_data.columns = ['Close','High','Low','Open','Volume']

appl_data['SMA_20'] = talib.SMA(appl_data['Close'].values, timeperiod=20)
# 2. 상대강도지수(RSI), 14일 기준
appl_data['RSI_14'] = talib.RSI(appl_data['Close'].values, timeperiod=13)

appl_data.tail(20)
```
